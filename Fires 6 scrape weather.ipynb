```python
import numpy as np
import pandas as pd
import requests
import json
import warnings
warnings.filterwarnings('ignore') # pd setting on a slice warnings ahead

def get_weather(row):
    try:
        fips = row['COUNTY_ID']
        date = row['DISCOVERY_DATE']
        payload = {'datasetid':'GHCND', 'datatypeid':['PRCP', 'TMAX', 'AWND'], 'locationid':'FIPS:%s' % fips, 'startdate':{date}, 'enddate':{date},'limit':'1000'}
        data = requests.get('https://www.ncdc.noaa.gov/cdo-web/api/v2/data', params=payload, headers={'token':'lkiVSiVEDTlPFfaLJSeIQaLxqWtPLzCo'})
        #print(data.url)
        prcps = []
        tmaxs = []
        awnd = []
        for dt, value in ((d['datatype'], d['value']) for d in data.json()['results']): 
            if dt == 'PRCP':
                prcps.append(value)
            elif dt == 'TMAX':
                tmaxs.append(value)
            elif dt == 'AWND':
                awnd.append(value)      
        return np.mean(prcps), np.mean(tmaxs), np.mean(awnd)
    
    except json.decoder.JSONDecodeError: # if the requests fails and data is not a json
        return -999, -999, -999 
    except KeyError: # if json is missing the datatypes 
        return -400, -400, -400
```
